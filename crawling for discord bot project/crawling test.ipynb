```python
import requests
from bs4 import BeautifulSoup
from info import url
```

```python
def get_html(url):
    response = requests.get(url)
    if response.status_code == 200:
        return response.text
    else:
        return None
```

```python
get_html(url["board_url"])
```

```python
soup = BeautifulSoup(get_html(url["board_url"]), 'html.parser')  
notices = []
rows = soup.find_all("tr")
rows
```

```python
for row in rows:
    number = row.find('td', class_ = "b-num-box")
    title =row.find('div', class_ = "b-title-box")
    notice_data = {'번호': number.get_text(strip=True) if number else '', '제목': title.a.get_text(strip=True) if title and title.a else ''}
    if notice_data["번호"] != '공지':
        notices.append(notice_data)
notices
```
